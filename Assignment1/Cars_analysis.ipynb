### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
avg_mpg=df["mpg"].mean()
df_sub=df[df["mpg"]>avg_mpg]
max_hp_wt=(df_sub["horsepower"]/df_sub["weight"]).max()
df_best_cars=list(df_sub.loc[(df_sub["horsepower"]/df_sub["weight"])==max_hp_wt,"name"])
print(df_best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
df_usa=df[df["origin"]=="usa"]
df_japan=df[df["origin"]=="japan"]
df_europe=df[df["origin"]=="europe"]
df_usa_1=df_usa.groupby(["model_year"])["mpg"].mean().reset_index(name="avg_usa_mpg")
df_japan_1=df_japan.groupby(["model_year"])["mpg"].mean().reset_index(name="avg_japan_mpg")
df_europe_1=df_europe.groupby(["model_year"])["mpg"].mean().reset_index(name="avg_europe_mpg")
plt.figure()
plt.plot(df_usa_1["model_year"],df_usa_1["avg_usa_mpg"],color="blue",label="USA")
plt.plot(df_japan_1["model_year"],df_japan_1["avg_japan_mpg"],color="green",label="Japan")
plt.plot(df_europe_1["model_year"],df_europe_1["avg_europe_mpg"],color="yellow",label="Europe")
plt.title("Evolution of average mpg over the years for each origin")
plt.ylabel("Average mpg")
plt.xlabel("Model year")
plt.legend()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
print(df.isna().sum())
sns.boxplot(y="horsepower",data=df)
plt.plot()
df["horsepower"]=df["horsepower"].fillna(df["horsepower"].median())

plt.figure(figsize=(8,9))
sns.scatterplot(
    data=df,
    x="horsepower",
    y="weight",
    hue="origin",
    hue_order=["japan","europe","usa"],
    palette="deep",
    size="mpg",
    sizes=(20,200),
    edgecolor="black",
    alpha=0.75,
    legend="brief"
)

plt.title("Horsepower by weight")
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
years_cars=df.groupby("name")["model_year"].nunique().reset_index(name="years_unique_count")
years_cars["average_mpg"]=df.groupby("name")["mpg"].mean().reset_index(name="average_mpg")["average_mpg"]
print(years_cars)#4thQuestion
print(years_cars.sort_values(["years_unique_count","average_mpg"],ascending=[0,0]))#Bonus question
cars_model_more_one=years_cars[years_cars["years_unique_count"]>1]["name"]
print(cars_model_more_one)#1st Question
cars_std=df[df["name"].isin(list(cars_model_more_one))].groupby("name")["mpg"].std().reset_index(name="mpg_std")
print(cars_std)#2nd Question
cars_model_more_one1=years_cars[years_cars["years_unique_count"]>1]
cars_merged=pd.merge(cars_model_more_one1,cars_std,on="name",how="inner")
mpg_std_min=cars_merged["mpg_std"].min()
cars_merged1=cars_merged[(cars_merged["mpg_std"]<1.0) & (cars_merged["years_unique_count"]>=2)]
print(cars_merged1[cars_merged1["mpg_std"]==mpg_std_min]["name"])#3rd Question



```
